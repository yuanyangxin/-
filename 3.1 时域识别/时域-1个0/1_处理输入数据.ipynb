```python
import os
import scipy.io as sio
import numpy as np
from shutil import copyfile
```

```python
path = "/workspace/PytorchCnn/Data/Signal0-30dB"
```

```python
files = os.listdir(path)
data = []
labels = []
```

```python
def del_file(filepath):
    """
    删除某一目录下的所有文件或文件夹
    :param filepath: 路径
    :return:
    """
    import shutil
    del_list = os.listdir(filepath)
    for f in del_list:
        file_path = os.path.join(filepath, f)
        if os.path.isfile(file_path):
            os.remove(file_path)
        elif os.path.isdir(file_path):
            shutil.rmtree(file_path)
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
saveFolder = "/workspace/PytorchCnn/Data/后面的稳定0信号/30dB/"
del_file(saveFolder)
for file in files:
#     print(file)
    device = file.split('_')[2]
    y = int(device) - 1
    round = int(file.split('_')[3][0:-4])
    data = sio.loadmat(os.path.join(path, file))
    X = data["y"]
    a = X[0: len(X)]
    
    real = a.real 
    imag = a.imag
    
    real = real.transpose()
    imag = imag.transpose()
    real = real[0]
    imag = imag[0]
#     print(real)
#     print(len(real))
    
#     print(imag)
#     print(len(imag))

    result = real[0: 160]
    im = imag[0: 160]
    result = np.vstack((result, im))

    name = "30dBProcess_" + str("A_" + device) + "_" + str(round) + ".npy"
    path1 = saveFolder + name
    np.save(path1, result)
```
