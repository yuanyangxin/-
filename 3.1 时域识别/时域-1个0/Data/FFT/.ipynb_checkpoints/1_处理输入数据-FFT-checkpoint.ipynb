step1: 对输入的I/O数据进行处理，I/O数据看作是nlp里面的文本数据

```python
import os
import scipy.io as sio
import numpy as np
from shutil import copyfile
```

```python
path = "/workspace/PytorchCnn/Data/FFT/30dB/"
```

```python
files = os.listdir(path)
data = []
labels = []
```

```python
def del_file(filepath):
    """
    删除某一目录下的所有文件或文件夹
    :param filepath: 路径
    :return:
    """
    import shutil
    del_list = os.listdir(filepath)
    for f in del_list:
        file_path = os.path.join(filepath, f)
        if os.path.isfile(file_path):
            os.remove(file_path)
        elif os.path.isdir(file_path):
            shutil.rmtree(file_path)
```

```python
saveFolder = "/workspace/PytorchCnn/Data/FFT/FFT-Process-30dB-60-100/"
del_file(saveFolder)
for file in files:
#     print(file)
    device = file.split('_')[2]
    y = int(device) - 1
    round = int(file.split('_')[3][0: -4])
    data = sio.loadmat(os.path.join(path, file))
    X = data["y1"]
    X = X[60:100]

    name = "30dBFFT_" + str("A_" + device) + "_" + str(round) + ".npy"
    path1 = saveFolder + name
    np.save(path1, X)
```
