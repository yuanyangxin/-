step2:按照训练集和测试集切割所有的数据

```python
import os
import random
```

```python
dir_path = "/workspace/GenratePic/Data_Process/"

targetFiles = sorted(os.listdir(dir_path))
flag = 0
```

```python
# 按照[文件名 标签名]的格式写入txt文件
with open(f'./files.txt', 'w') as f:
    for targetFile in targetFiles:
        # print(targetFile)
        label = int(targetFile.split("_")[2]) - 1
        f.write(f'{os.path.join(dir_path, targetFile)} {label}\n')
```

```python
# 分离训练集和测试集
with open(f'./files.txt', 'r') as f, open(f'./files_train.txt', 'w') as f_train, open(f'./files_test.txt',
                                                                                      'w') as f_test:
    lines = f.readlines()
    for line in lines:
        print(line)
        filepath = line.split(" ")[0]
        print(filepath)
        filename = filepath.split("/")[4]
#         print(filename)
        round = int(filename.split("_")[3][:-4])
#         print(round)
        if 1<= round <= 4:
            f_test.write(line)
        else:
            f_train.write(line)
```
