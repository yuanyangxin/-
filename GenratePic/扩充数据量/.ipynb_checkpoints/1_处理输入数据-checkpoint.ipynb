step1: 对输入的I/O数据进行处理，I/O数据看作是nlp里面的文本数据

```python
import os
import scipy.io as sio
import numpy as np
from shutil import copyfile
```

```python
path = "/workspace/GenratePic/Data_Set/"
```

```python
files = os.listdir(path)
data = []
labels = []
```

```python
def del_file(filepath):
    """
    删除某一目录下的所有文件或文件夹
    :param filepath: 路径
    :return:
    """
    import shutil
    del_list = os.listdir(filepath)
    for f in del_list:
        file_path = os.path.join(filepath, f)
        if os.path.isfile(file_path):
            os.remove(file_path)
        elif os.path.isdir(file_path):
            shutil.rmtree(file_path)
```

```python
# 1. 查找Count_Data_Length_Sides，找出54台设备中的最小距离 和 每组数据中的实验数量
minDistance = float('inf')
numInEachData = -1
for file in files:
    device = file.split('_')[2]
    y = int(device) - 1
    round = int(file.split('_')[4][0:-4])
#     print(round)
    data = sio.loadmat(os.path.join(path, file))
    lengths = data["Count_Data_Length_Sides"]
#     print(lengths)
#     print(lengths.shape)
    if numInEachData == -1:
        numInEachData = len(lengths)
    if len(lengths) != numInEachData:
        print("Warning! Error")
        print(len(lengths))
        print(file)
    for length in lengths:
        minDistance = min(minDistance, length[1] - length[0] + 1)
print(minDistance)
print(numInEachData)
```

```python
#2. 确定最接近的size的大小 
import math
size = math.floor(math.sqrt(minDistance))
print(size)
```

```python
# 3. 截取数据
%matplotlib inline
indexNum = size * size // 2
import matplotlib.pyplot as plt
saveFolder = "/workspace/GenratePic/扩充数据量/Data_Process/"
del_file(saveFolder)

for file in files:
    device = file.split('_')[2]
    y = int(device) - 1
    round = int(file.split('_')[4][0:-4])
#     print(round)
    data = sio.loadmat(os.path.join(path, file))
    lengths = data["Count_Data_Length_Sides"]
    X = data["Brush_Data_Temp"]
    
    for i in range(len(lengths)):
        term = numInEachData * (round - 1) + (i + 1)
        print(term)
        start = lengths[i][0]
        end = lengths[i][1]
        a = X[0: indexNum]
    
        real = a.real 
        imag = a.imag

        real = real.transpose()
        imag = imag.transpose()
        real = real[0]
        imag = imag[0]
    #     print(real)
    #     print(len(real))

    #     print(imag)
    #     print(len(imag))

        result = real[0: len(X)]
        im = imag[0: len(X)]
        result = np.vstack((result, im))
        result = result.reshape((size, size))

        name = "30dBProcess_" + str("A_" + device) + "_" + str(term) + ".npy"
        path1 = saveFolder + name
        np.save(path1, result)
```
